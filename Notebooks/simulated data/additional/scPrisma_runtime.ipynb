```python
import os
from ipylab import JupyterFrontEnd
from algorithms_torch import *
from data_gen import *
from datasets import *
import time
import torch
torch.cuda.get_device_name()
```

```python
ncells =[]
time_recon = []
for i in range(18):
    ncells.append(1000*(i+1))
    A = simulate_spatial_cyclic(2000,1000*(i+1),w=0.3)
    A = A + np.random.normal(0,0.1,A.shape)
    t = time.time()
    E , E_recon  = reconstruction_cyclic_torch(A, iterNum=100 , verbose=False)
    elapsed = time.time() - t
    time_recon.append(elapsed)
    del A
    del E
    del E_recon
    torch.cuda.empty_cache()
    print("Number of cells: " +str(1000*(i+1)) +" elapsed time: " +str(elapsed))

ncells = np.array(ncells)
time_recon = np.array(time_recon)

with open('time_recon.npy', 'wb') as f:
    np.save(f, time_recon)
with open('ncells_recon.npy', 'wb') as f:
    np.save(f, ncells)
app = JupyterFrontEnd()
app.commands.execute('docmanager:save')
```

```python
ncells_filtering =[]
time_filtering = []
for i in range(17):
    ncells_filtering.append(2000*(i+1))
    A = simulate_spatial_cyclic(2000,2000*(i+1),w=0.3)
    A = A + np.random.normal(0,0.1,A.shape)
    t = time.time()
    F  = filtering_cyclic_torch(A, iterNum=100)
    elapsed = time.time() - t
    time_filtering.append(elapsed)
    torch.cuda.empty_cache()
    del A
    del F
    torch.cuda.empty_cache()
    print("Number of cells: " +str(2000*(i+1)) +" elapsed time: " +str(elapsed))
ncells_filtering = np.array(ncells_filtering)
time_filtering = np.array(time_filtering)

with open('time_filtering.npy', 'wb') as f:
    np.save(f, time_filtering)
with open('ncells_filtering.npy', 'wb') as f:
    np.save(f, ncells_filtering)
app = JupyterFrontEnd()
app.commands.execute('docmanager:save')
```

```python
ncells_genes =[]
time_genes = []
for i in range(17):
    ncells_genes.append(2000*(i+1))
    A = simulate_spatial_cyclic(2000,2000*(i+1),w=0.3)
    A = A + np.random.normal(0,0.1,A.shape)
    t = time.time()
    D  = filter_non_cyclic_genes_torch(A, iterNum=100)
    elapsed = time.time() - t
    time_genes.append(elapsed)
    del A
    del D
    torch.cuda.empty_cache()
    print("Number of cells: " +str(2000*(i+1)) +" elapsed time: " +str(elapsed))
ncells_genes = np.array(ncells_genes)
time_genes = np.array(time_genes)

with open('time_genes.npy', 'wb') as f:
    np.save(f, time_genes)
with open('ncells_genes.npy', 'wb') as f:
    np.save(f, ncells_genes)
app = JupyterFrontEnd()
app.commands.execute('docmanager:save')
```
