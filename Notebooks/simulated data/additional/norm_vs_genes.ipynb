```python
from scipy.stats import entropy , pearsonr
from algorithms import *
from datasets import *
from data_gen import *
import scanpy as sc
import numpy as np
```

```python
B = simulate_spatial_cyclic(100,100,w=0.3)
noise = np.random.normal(0,0.5,(100,100))
noise = np.clip(noise,0,np.inf)
norm_list = []
ngenes_list = []
noise_norm = []
for i in range(100):
    print('($$$$)')
    print(i)
    print(norm_list)
    print('($$$$)')
    A = noise.copy()
    A[:,:i] = B[:,:i] + A[:,:i]
    A = A/np.linalg.norm(A)
    np.random.shuffle(A)
    E , E_rec = reconstruction_cyclic(A , iterNum=100 , verbose=False)
    A_orig= A.copy()
    A = E_rec @ A
    D = filter_cyclic_genes(A, regu=2 , iterNum=100)
    D = np.identity(D.shape[0]) - D
    A = A @ D
    F = enhancement_cyclic(A , regu=0.2 , iterNum=100 , verbosity=200)
    A = A * F
    noise_norm.append(np.linalg.norm(noise))
    ngenes_list.append(i)
    norm_list.append(np.linalg.norm(A))
```

```python
fig,ax = plt.subplots()
# make a plot
ax.plot(ngenes_list,
        norm_list,
        color="red", 
        marker="o")
# set x-axis label
ax.set_xlabel("Number of cyclic genes", fontsize = 16)
# set y-axis label
ax.set_ylabel("Norm after applying scPrisma",
              
              fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=15)
ax.tick_params(axis='both', which='minor', labelsize=15)
#plt.title()
plt.show()
# save the plot as a file
fig.savefig('two_different_y_axis_for_single_python_plot_with_twinx.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
```
