```python
import copy

import matplotlib.pyplot as plt
import numpy as np
import scanpy as sc
import pandas as pd
from algorithms import *
from data_gen import *
from spectrum_gen import *
from visualizations import *
from analysis import *
import random

```

```python
noise_list =[]

corr_list = []
snr_list = []
corr_list_in = []

T = simulate_spatial_cyclic(ngenes=500,ncells=99,w=0.3)
t_norm = np.linalg.norm(T)
indicator = np.zeros((99,99))
for i in range(33):
    for j in range(33):
        indicator[i,j]=1
        indicator[i+33,j+33]=1
        indicator[i+66,j+66]=1
    
for i in range(1000):
        print("Iteration number: " + str(i))
        noise = np.random.normal(0,0.2 + i*0.01,(T.shape))
        noise = np.clip(noise,0,np.inf)
        K = T + noise
        E_sga  , E_rec_sga = sga_m_reorder_rows_matrix(K,iterNum=100,batch_size=90)
        res = spearm(E_rec_sga,np.array(range(K.shape[0])))
        corr_list.append(res)
        noise_list.append(0.2 + i*0.01)
        print(res)
        noise_norm = np.linalg.norm(noise)
        snr = t_norm/noise_norm
        snr_list.append(snr)
        E_sga  , E_rec_sga = reorder_indicator(K,indicator,iterNum=100,batch_size=90)
        res = spearm(E_rec_sga,np.array(range(K.shape[0])))
        corr_list_in.append(res)
```

```python
fig, ax = plt.subplots()

ax.plot((np.array(snr_list)),(np.array(corr_list)), 'bo',color='b' , label="No use of prior knowledge")
ax.plot( (np.array(snr_list)),savgol_filter(np.array((np.array(corr_list))),97,3), 'k', linewidth=3.0 )#, label="No use of prior knowledge smoothed")
ax.plot((np.array(snr_list)),(np.array(corr_list_in)), 'bo',color='g' , label="Use prior knowledge of 3 bins")
ax.plot( (np.array(snr_list)),savgol_filter(np.array((np.array(corr_list_in))),97,3), 'k', linewidth=3.0)# , label="Use prior knowledge smoothed")
plt.legend()
ax.set_xlabel("SNR ",  fontsize=16)
ax.set_ylabel("Spearman correlation ",  fontsize=16)
ax.set_title("Spearman correlation as function of the SNR ",  fontsize=16)
ax.grid(True)
ax.invert_xaxis()
plt.tick_params(labelsize=16)
plt.ylim(0, 1.1)

plt.show()

plt.show()
```
