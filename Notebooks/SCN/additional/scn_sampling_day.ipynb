```python
import scanpy as sc
from visualizations import *
from datasets import *
```

```python
cluster_names = [
    'SCN Neurons', 'Oligodendrocytes','Ependymal',
    'NG2','Endothelial', 
    'Astrocytes',
    'Microglia', 'Tanycytes']

for cluster in cluster_names:
    adata = sc.read("SCN/" + cluster + "_en.h5ad")
    adata.obs["day"]=(np.array(adata.obs["ST"]).astype('float64')  //38).astype('str')
    sc.tl.pca(adata)
    sc.pl.pca_scatter(adata, color='CT' , title=("PCA of enhanced "  + cluster + ", painted by circadian time"), legend_fontsize='x-large')
    sc.pl.pca_scatter(adata, color='day' , title=("PCA of enhanced "  + cluster + ", painted by sampling day"), legend_fontsize='x-large')
```
